## Exploration

```python
df = pd.read_excel(pruned_aricles_armed_conflict_path)
```

```python
df.columns
```

```python
articles_civilian = 0
articles_non_civilian = 0

for index, row in tqdm(df.iterrows()):
    headline = row.source_headline
    if 'pacifist' in headline or 'civilians' in headline or 'civilian' in headline or 'civil' in headline:
        articles_civilian += 1
    else:
        articles_non_civilian += 1
```

```python
print(articles_civilian)
print(articles_non_civilian)
```

```python
deaths_by_side = {'Civilians' : 0}

for index, row in tqdm(df.iterrows()):
    if row.side_a not in deaths_by_side.keys():
        deaths_by_side[row.side_a] = 0
    if row.side_b not in deaths_by_side.keys():
        deaths_by_side[row.side_b] = 0
        
    deaths_by_side[row.side_a] += row.deaths_a
    deaths_by_side[row.side_b] += row.deaths_b
    deaths_by_side['Civilians'] += row.deaths_civilians
```

```python
sorted_death_by_side = sorted(deaths_by_side.items(), key=operator.itemgetter(1), reverse=True)
```

```python
zone_dict = {}
for index, row in df.iterrows():
    zone = row.where_coordinates
    
    if zone not in zone_dict.keys():
        zone_dict[zone] = 0
        
    zone_dict[zone] += 1
```

```python
sorted_zone = sorted(zone_dict.items(), key=operator.itemgetter(1), reverse=True)
```
