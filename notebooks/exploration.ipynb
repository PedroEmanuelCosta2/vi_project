## Exploration

```python
armed_conflict_df = pd.read_excel(armed_conflict_path)
```

```python
armed_conflict_df.columns
```

```python
source_armed_conflict_df = armed_conflict_df[['id', 'year', 'date_start', 'date_end',
                                              'number_of_sources', 'source_article', 
                                              'source_office', 'source_date', 'source_headline',
                                              'source_original']]
```

```python
articles_civilian = 0
articles_non_civilian = 0

for index, row in tqdm(source_armed_conflict_df.dropna(subset=['source_headline']).iterrows()):
    headline = row.source_headline
    if 'pacifist' in headline or 'civilians' in headline or 'civilian' in headline or 'civil' in headline:
        articles_civilian += 1
    else:
        articles_non_civilian += 1
```

```python
print(articles_civilian)
print(articles_non_civilian)
```

```python
pruned_armed_conflict_df = armed_conflict_df[['id', 'year', 'date_start', 'date_end', 
                                              'conflict_name', 'side_a', 'side_b', 'latitude', 
                                              'longitude', 'where_coordinates','country', 'country_id', 
                                              'deaths_a', 'deaths_b', 'deaths_civilians']]
```

```python
deaths_by_side = {'Civilians' : 0}

for index, row in tqdm(pruned_armed_conflict_df.iterrows()):
    if row.side_a not in deaths_by_side.keys():
        deaths_by_side[row.side_a] = 0
    if row.side_b not in deaths_by_side.keys():
        deaths_by_side[row.side_b] = 0
        
    deaths_by_side[row.side_a] += row.deaths_a
    deaths_by_side[row.side_b] += row.deaths_b
    deaths_by_side['Civilians'] += row.deaths_civilians
```

```python
sorted_death_by_side = sorted(deaths_by_side.items(), key=operator.itemgetter(1), reverse=True)
```
